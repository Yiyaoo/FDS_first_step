```python
import pandas as pd
import numpy as np
import io
import glob
import requests
```

```python
#read files from github
url="https://raw.githubusercontent.com/Yiyaoo/FDS_first_step/master/FDS_Fourrage/FDS_FOURRAGE_2020.csv"
s=requests.get(url).content
df=pd.read_csv(io.StringIO(s.decode('utf-8')),sep=';')
```

```python
#df = df.reset_index()
count=0
df_suf = pd.DataFrame()
for index, row in df.iterrows():
  if (row.DEP!='...')&(row.N027_MOD==70.02)&(pd.isna(row.VALEUR)):
    #print(row.REGION,row.DEP,row.N306_MOD,row.VALEUR)
    #count = count+1
    df_superficie = df.loc[(df.REGION == row.REGION) & (df.DEP == row.DEP) & (df.N306_MOD==row.N306_MOD) & (df.N027_MOD==50.07)]
    df_production = df.loc[(df.REGION == row.REGION) & (df.DEP == row.DEP) & (df.N306_MOD==row.N306_MOD) & (df.N027_MOD==20.08)]
    if int(df_superficie.VALEUR)!=0:
      rendement = int(df_production.VALEUR)/int(df_superficie.VALEUR)
    else:
      rendement =0
    df.loc[[index],'VALEUR'] = rendement
    #print(rendement)
```

```python
df.to_csv(sep = ';')
```
