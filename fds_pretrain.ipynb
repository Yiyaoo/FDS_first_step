```python
import pandas as pd
import numpy as np
import io
import glob
import requests
```

**Get Data From Github**

```python
#read files from github
url="https://raw.githubusercontent.com/Yiyaoo/FDS_first_step/master/FDS_Fourrage/FDS_FOURRAGE_2020.csv"
s=requests.get(url).content
df=pd.read_csv(io.StringIO(s.decode('utf-8')),sep=';')
```

```python
#Données Developee
#add missing rendement values
#df = df.reset_index()
count=0
df_suf = pd.DataFrame()
for index, row in df.iterrows():
  if (row.REGION!='...')&(row.DEP!=None)&(row.N027_MOD==70.01)&(pd.isna(row.VALEUR)):
    #print(row.REGION,row.DEP,row.N306_MOD,row.VALEUR)
    count = count+1
    df_superficie = df.loc[(df.REGION == row.REGION) & (df.DEP == row.DEP) & (df.N306_MOD==row.N306_MOD) & (df.N027_MOD==50.08)]
    #print (df_superficie.VALEUR)
    df_production = df.loc[(df.REGION == row.REGION) & (df.DEP == row.DEP) & (df.N306_MOD==row.N306_MOD) & (df.N027_MOD==20.01)]
    if int(df_superficie.VALEUR)!=0:
      rendement = int(df_production.VALEUR)/int(df_superficie.VALEUR)
    else:
      rendement =0
    df.loc[[index],'VALEUR'] = rendement



#delete rendement=0 lignes
for index, row in df.iterrows():
  if (row.N027_MOD==70.01)&(row.VALEUR==0):
    count = count+1
    #print(index)
    df=df.drop(index)
```
