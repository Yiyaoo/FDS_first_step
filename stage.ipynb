```python
#read multiples files from github

def read_multi_csv(start_year,end_year):
    years = list(range(start_year,end_year+1))
    dfs = []
    for YYYY in years:
        #file = 'https://raw.githubusercontent.com/username/project/main/data/normalized/'+str(YYYY)+'_crimes_byState.csv'
        file = 'https://raw.githubusercontent.com/Yiyaoo/playground-DHnDOLo0/master/FDS_Fourrage/FDS_FOURRAGE_'+str(YYYY)+'.csv'
        #print (file)
        df = pd.read_csv(file,sep=';')
        dfs.append(df)
    all_dfs = pd.concat(dfs)
    return all_dfs

df=read_multi_csv(2000,2020)
```

```python
#données manquantes par type de culture et par année
df.drop(['ANNREF', 'N306_LIB'], 1).isna().groupby([df.ANNREF, df.N306_LIB], sort=False).sum().reset_index()
```
